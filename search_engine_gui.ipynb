## Search engine gui

```python
from tkinter import *
from PIL import ImageTk, Image
import webbrowser

# Create object 
root = Tk()
  
# Adjust size 
root.geometry("900x600")
  
# Add image file bcakground
bg = ImageTk.PhotoImage(Image.open("./images_gui/pexels-harry-cooke-6195041.png"))

#search icon text
img = Image.open("./images_gui/search1.png")
img_resized = img.resize((90,80), Image.ANTIALIAS)
search_icon_t = ImageTk.PhotoImage(img_resized)

#search icon image
img = Image.open("./images_gui/search2.png")
img_resized = img.resize((96,96), Image.ANTIALIAS)
search_icon_i = ImageTk.PhotoImage(img_resized)

#Github 
img = Image.open("./images_gui/github.png")
img_resized = img.resize((30,30), Image.ANTIALIAS)
github_icon_paresh = ImageTk.PhotoImage(img_resized)
github_icon_meenakshi = ImageTk.PhotoImage(img_resized)

#linkedin
img = Image.open("./images_gui/linkedinn.png")
img_resized = img.resize((30,30), Image.ANTIALIAS)
linkedin_icon_paresh = ImageTk.PhotoImage(img_resized)
linkedin_icon_meenakshi = ImageTk.PhotoImage(img_resized)

#button text
img = Image.open("./images_gui/button_text.png")
img_resized = img.resize((220,50), Image.ANTIALIAS)
button_text = ImageTk.PhotoImage(img_resized)

#button text
img = Image.open("./images_gui/button_image.png")
img_resized = img.resize((220,50), Image.ANTIALIAS)
button_image = ImageTk.PhotoImage(img_resized)

#button text
img = Image.open("./images_gui/button_both.png")
img_resized = img.resize((220,50), Image.ANTIALIAS)
button_both = ImageTk.PhotoImage(img_resized)

# Create Canvas
canvas1 = Canvas( root, width = 900, height = 520 ,bd =0 ,highlightthickness=0)
#canvas2 = Canvas( root, width = 900, height = 80,bd =0 ,highlightthickness=0)
  
canvas1.pack(side='top', fill = "both", expand = True)
#canvas2.pack(side='top',fill = "both", expand = True)

# Display image
canvas1.create_image( 0, 0, image = bg, anchor = "nw")

def callback(url):
    webbrowser.open_new(url)
    
def text_search_tfidf():
    canvas1.delete("word2vec")
    canvas1.delete("tfidf")
    word = text_enter.get("1.0",'end-1c')
    result = search_tfidf(str(word))
    canvas1.create_line(0, 120, 900, 120)
    for i in range(0,7):
        if i==0 or i==6:
            canvas1.create_text(150 , (i)*40 + 150, text = result[i] , fill='black', anchor=NW ,font=('Arapey','12','italic'),tag="tfidf")
        else:
            canvas1.create_text(150 , (i)*40 + 150, text = result[i] , fill='#472183', anchor=NW ,font=('Arapey','12','normal') ,tag="tfidf")
                
def text_search_word2vec():
    canvas1.delete("word2vec")
    canvas1.delete("tfidf")
    word = text_enter.get("1.0",'end-1c')
    result = search_word2vec(str(word))
    canvas1.create_line(0, 120, 900, 120)
    
    for i in range(0,8):
        if i==0 or i==1 or i==7:
            canvas1.create_text(150 , (i)*40 + 150, text = result[i] , fill='black', anchor=NW ,font=('Arapey','12','italic'), tag="word2vec")
        else:
            canvas1.create_text(150 , (i)*40 + 150, text = result[i] , fill='#472183', anchor=NW ,font=('Arapey','12','normal'), tag="word2vec")
    
def text_search():
    canvas1.delete("all")
    canvas1.configure(bg='#C0DEFF')
    canvas1.create_text( 210, 50, text = "Enter word here", fill='black',font=('Arapey','18','normal'))
    global text_enter
    text_enter = Text(root,height=1, width = 25 ,fg="black",font=('Arapey','15','normal'), bd=0)
    canvas1.create_window( 320,35, anchor=NW, window=text_enter)
    button_search_tfidf = Button(root, text="Search using tfidf", command=text_search_tfidf, borderwidth=0,font=('Arapey','10','normal'), cursor="hand2")
    canvas1.create_window( 330, 80 , anchor=NW, window=button_search_tfidf)
    
    button_search_word2vec = Button(root, text="Search using word2vec", command=text_search_word2vec, borderwidth=0,font=('Arapey','10','normal'), cursor="hand2" )
    canvas1.create_window( 470, 80 , anchor=NW, window=button_search_word2vec)

def image_search_bow():
    canvas1.delete("bow")
    word = text_enter.get("1.0",'end-1c')
    image_path = "./search_images/" + str(word)
    result = search_image(image_path)
    canvas1.create_line(0, 120, 900, 120)
    canvas1.create_text(150 , 150, text = "Top similar images : " , fill='black', anchor=NW ,font=('Arapey','12','italic') ,tag="bow")
    for i in range(0,3):
        print(i)
        iph = "./search_results/" + "result" + str(i+1) +".jpg"
        img = Image.open(iph)
        img_resized = img.resize((150,150), Image.ANTIALIAS)
        ri = ImageTk.PhotoImage(img)
        canvas1.create_image(  100 , 300 , anchor=NW,image=ri, tag="bow")
    
def image_search():
    canvas1.delete("all")
    canvas1.configure(bg='#C0DEFF')
    canvas1.create_text( 210, 50, text = "Enter image name", fill='black',font=('Arapey','18','normal'))
    global text_enter
    text_enter = Text(root,height=1, width = 25 ,fg="black",font=('Arapey','15','normal'), bd=0)
    canvas1.create_window( 320,35, anchor=NW, window=text_enter)
    button_search_tfidf = Button(root, text="Search", command=image_search_bow, borderwidth=0,font=('Arapey','10','normal'), cursor="hand2")
    canvas1.create_window( 330, 80 , anchor=NW, window=button_search_tfidf)
    
def multimodal_search():
    canvas1.delete("all")
    canvas1.configure(bg='#C0DEFF')
    canvas1.create_text( 210, 50, text = "Enter here", fill='black',font=('Arapey','18','normal'))
    global text_enter
    text_enter = Text(root,height=1, width = 25 ,fg="black",font=('Arapey','15','normal'), bd=0)
    canvas1.create_window( 320,35, anchor=NW, window=text_enter)
    
    def clicked(value):
        canvas1.delete("mm")
        word = text_enter.get("1.0",'end-1c')
        canvas1.create_line(0, 120, 900, 120)
        if value == "text" :
            mm_text = multimodal_search_word(word)
            for i in range(0,8):
                if i==0 or i==1 or i==7:
                    canvas1.create_text(150 , (i)*40 + 150, text = mm_text[i] , fill='black', anchor=NW ,font=('Arapey','12','italic'), tag="mm")
                else:
                    canvas1.create_text(150 , (i)*40 + 150, text = mm_text[i] , fill='#472183', anchor=NW ,font=('Arapey','12','normal'), tag="mm")
        else:
            image_path = "./search_images/" + str(word)
            mm_image = multimodal_search_image(image_path)
            for i in range(0,8):
                if i==0 or i==1 or i==7:
                    canvas1.create_text(150 , (i)*40 + 150, text = mm_image[i] , fill='black', anchor=NW ,font=('Arapey','12','italic'), tag="mm")
                else:
                    canvas1.create_text(150 , (i)*40 + 150, text = mm_image[i] , fill='#472183', anchor=NW ,font=('Arapey','12','normal'), tag="mm")
        
   
    TOPPINGS = [
        ("text", "text"),
        ("image", "image"),
    ]
    
    option = StringVar()
    option.set("text")
    x = 370
    
    for text, topping in TOPPINGS:
        radio = Radiobutton(root, text=text, variable=option, value=topping, bg='#C0DEFF' , borderwidth=0,font=('Arapey','10','normal'), cursor="hand2")
        canvas1.create_window(x, 80 , anchor=NW, window=radio)
        x=x+100
        
    button_search_tfidf = Button(root, text="Search", command=lambda: clicked(option.get()), borderwidth=0,font=('Arapey','10','normal'), cursor="hand2")
    canvas1.create_window( 600, 35 , anchor=NW, window=button_search_tfidf)
     
# Create Buttons
button_text1 = Button( root, image=button_text, command=text_search, borderwidth=0,cursor="hand2")
button_image2 = Button( root, image=button_image, command=image_search, borderwidth=0 , cursor="hand2")
button_both3 = Button( root, image=button_both, command=multimodal_search, borderwidth=0 , cursor="hand2" )
#button_linkedin_m = Button( root, image=linkedin_icon_meenakshi, command=callback("http://www.google.com"), borderwidth=0, cursor="hand2")
#button_linkedin_p = Button( root, image=linkedin_icon_paresh, command=callback("http://www.google.com"), borderwidth=0, cursor="hand2")
#button_github_m = Button( root, image=github_icon_meenakshi, command=callback("http://www.google.com"), borderwidth=0, cursor="hand2")
#button_github_p = Button( root, image=github_icon_paresh, command=callback("http://www.google.com"), borderwidth=0, cursor="hand2")


# Add Text
canva_searchmessage1 = canvas1.create_text( 80, 120,anchor=NW, text = "MultiModal Search Engine",fill='white',font=('Arapey','35','normal'), tag="first_window")
canva_searchmessage2 = canvas1.create_text( 80, 200, anchor=NW, text = "Search related words, images or both!!!",fill='white',font=('Arapey','15','normal'), tag="first_window")
canva_image6 = canvas1.create_window( 580, 200 , anchor=NW,window=button_text1, tag="first_window")
canva_image7 = canvas1.create_window( 580, 275 , anchor=NW,window=button_image2, tag="first_window")
canva_image8 = canvas1.create_window( 580, 345 , anchor=NW,window=button_both3, tag="first_window")
canva_image1 = canvas1.create_image( 300, 250 , anchor=NW,image=search_icon_t, tag="first_window")
canva_image2 = canvas1.create_image( 150, 250 , anchor=NW,image=search_icon_i, tag="first_window")

c_message = Label(root, text="Developed by Meenakshi and Paresh", fg="black", font=('Arapey',10,'italic'))
c_message.pack(pady=4)

c_message = Label(root, text="         ")
c_message.pack(padx=156, side=LEFT)


button_github_m = Label(root, image=github_icon_meenakshi, cursor="hand2")
button_github_m.pack(padx=10, side=LEFT)
button_github_m.bind("<Button-1>", lambda e: callback("https://github.com/Meenakshi-Meena"))

button_linkedin_m = Label(root, image=linkedin_icon_meenakshi , cursor="hand2")
button_linkedin_m.pack(padx=10, side=LEFT)
button_linkedin_m.bind("<Button-1>", lambda e: callback("https://www.linkedin.com/in/meenakshi-meena-9a8033228"))

button_github_p = Label(root, image=github_icon_paresh , cursor="hand2")
button_github_p.pack(padx=10, side=LEFT)
button_github_p.bind("<Button-1>", lambda e: callback("https://github.com/paresh-bhagat"))

button_linkedin_p = Label(root, image=linkedin_icon_paresh , cursor="hand2")
button_linkedin_p.pack(padx=10, side=LEFT)
button_linkedin_p.bind("<Button-1>", lambda e: callback("https://www.linkedin.com/in/paresh-bhagat-1a9154148"))


#canva_searchmessage3 = canvas2.create_text( 350,10, anchor=NW, text = "Created by Meenakshi and Paresh",fill='black',font=('Arapey','10','italic'), tag="first_window")
#canva_image2 = canvas2.create_window( 470, 35 , anchor=NW,window=button_github_p, tag="first_window")
#canva_image3 = canvas2.create_window( 390, 35 , anchor=NW,window=button_github_m, tag="first_window")
#canva_image4 = canvas2.create_window( 500, 35 , anchor=NW,window=button_linkedin_p, tag="first_window")
#canva_image5 = canvas2.create_window( 420, 35 , anchor=NW,window=button_linkedin_m , tag="first_window")

# Execute tkinter
root.mainloop()
```
